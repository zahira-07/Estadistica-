```python
import pandas as pd
import numpy as np
#subir la base de datos
df=pd.read_csv("https://raw.githubusercontent.com/zahira-07/Estadistica-/refs/heads/main/Wine%20dataset.csv")
df.dropna(inplace=True)
df.drop(columns=["Flavanoids","Proanthocyanins","Proanthocyanins"],inplace=True)
df.corr()
```

encontrar la correlacion entre la dependencia del alcohol, en como influye en el color del vino.

```python
#1 Establesca una variable dependiente ( Y ) y una variable independiente ( X ).
# independiente ( x ).

# variable independiente:class
# variable dependiente:Color intensity
import numpy as np
x = df['Color intensity']
y = df['Alcohol']
```

: La relación entre el contenido de alcohol y el color del vino es un tema de interés en la industria del vino. El alcohol es un componente importante del vino que puede influir en la estabilidad y la intensidad del color.
La correlación lineal es una medida estadística que se utiliza para evaluar la relación entre dos variables. En este caso, se busca evaluar la relación entre el contenido de alcohol y el color del vino.


```python
# 2. Realice un diagrama de dispersión para estos datos.
import matplotlib.pyplot as plt
plt.scatter(x, y, color = 'red')
plt.xlabel('Alcohol') # nombrar eje x
plt.ylabel('Color intensity') # nombrar eje y
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

El diagrama de dispersión es una herramienta visual que nos permite analizar la relación entre dos variables. En este caso, hemos utilizado un diagrama de dispersión para evaluar la relación entre el contenido de alcohol y el color del vino.

Al observar el diagrama de dispersión, podemos ver que hay una tendencia positiva en la relación entre el contenido de alcohol y el color del vino. Esto significa que a medida que aumenta el contenido de alcohol, el color del vino también tiende a aumentar.

El diagrama de dispersión también nos muestra que la relación entre el contenido de alcohol y el color del vino no es perfectamente lineal. Hay algunos puntos que se encuentran por encima o por debajo de la línea de tendencia, lo que sugiere que hay otros factores que también influyen en la relación entre el contenido de alcohol y el color del vino.

Sin embargo, en general, el diagrama de dispersión sugiere que el contenido de alcohol es un factor importante que influye en el color del vino. Esto es consistente con nuestra intuición, ya que el alcohol es un componente importante del vino que puede afectar su color y su sabor.

```python
# 3. Calcule el coeficiente de correlación e interprete el resultado.
Sxx = np.sum(   (x- np.mean(x)) ** 2    )
Sxy = np.sum(    (y - np.mean(y))  * (x - np.mean(x)))
Syy = np.sum(   (y - np.mean(y)   )   **   2)
```

Con un coeficiente de correlación (r) de 0.54, podemos decir que hay una correlación moderada entre el contenido de alcohol y el color del vino.
El coeficiente de correlación mide la fuerza y la dirección de la relación lineal entre dos variables. En este caso, un valor de 0.54 indica que:
- La relación es positiva, es decir, a medida que aumenta el contenido de alcohol, el color del vino también tiende a aumentar.
- La relación es moderada, lo que significa que no es muy fuerte ni muy débil.


```python
# 4.Calcule el coeficiente de determinación e interprete el resultado.
r = Sxy / np.sqrt(Sxx * Syy)
print('Coeficiente de correlacion:  ',  r)
print('coeficiente de determinacion:  ',  r** 2)
```

Con un coeficiente de determinación (R²) de 0.2985, podemos decir que aproximadamente el 29.85% de la varianza en el color del vino se explica por el contenido de alcohol.
En otras palabras, el modelo de regresión lineal que relaciona el contenido de alcohol con el color del vino explica aproximadamente un tercio de la variabilidad en el color del vino.
Esto significa que hay otros factores que no se han considerado en el modelo que también influyen en el color del vino.
Es importante destacar que el coeficiente de determinación no mide la bondad del ajuste del modelo, sino más bien la proporción de la varianza explicada por el modelo.
En resumen, con un R² de 0.2985, podemos decir que el contenido de alcohol explica aproximadamente un tercio de la variabilidad en el color del vino, pero que hay otros factores que también influyen en el color del vino


```python
# 5. Obtenga la recta de regresión ajustada y grafíquelo sobre el gráfico de dispersión.
b1 = Sxy / Sxx
b0 = np.mean(y) - b1 * np.mean(x)

fun = lambda x: b1 * x + b0  # crea una función para la regresión lineal ajustada

Yc = fun(x)

plt.plot(x, Yc, color = 'black', linestyle = '--')
```

La recta de regresión ajustada revela una relación significativa entre el contenido de alcohol y el color del vino. Los resultados sugieren que por cada unidad de aumento en el contenido de alcohol, el color del vino aumenta en aproximadamente 0.19 unidades.

Esta relación positiva entre el contenido de alcohol y el color del vino sugiere que el alcohol juega un papel importante en la determinación del color del vino. Es probable que el alcohol influya en la formación de compuestos fenólicos y antocianinas, que son responsables del color del vino.

La recta de regresión ajustada también sugiere que otros factores, como la variedad de uva, la región de origen y el proceso de elaboración, pueden influir en la relación entre el contenido de alcohol y el color del vino.

En resumen, la recta de regresión ajustada revela una relación significativa entre el contenido de alcohol y el color del vino, lo que sugiere que el alcohol es un factor importante en la determinación del color del vino.

```python
# 6. Obtenga un intervalo de confianza del 95% para la pendiente de la recta de regresión ajustada ( b1 )
nivel_de_significancia = 0.05
from scipy.stats import t
T = t.ppf(1 - nivel_de_significancia / 2, df = x.size - 2)
se_b1 = np.sqrt(np.sum((y - Yc) ** 2) / (y.size - 2)) / np.sqrt(Sxx)
confianza_b1 = b1 - T * se_b1, b1 + T * se_b1
print(f'intervalo de confianza de {1 - nivel_de_significancia: 0.0%}')
print(f'{confianza_b1[0] : 0.4f} < b1 < {confianza_b1[1] : 0.4f} \n')
```

El intervalo de confianza de 95% que has obtenido es:
0.1477 < b1 < 0.2350
Donde b1 es el coeficiente de regresión que relaciona el contenido de alcohol con el color del vino.
Este intervalo de confianza nos dice que, con un nivel de confianza del 95%, podemos afirmar que el verdadero valor del coeficiente de regresión b1 se encuentra dentro del rango de 0.1477 a 0.2350.
En otras palabras, podemos estar seguros al 95% de que, por cada unidad de aumento en el contenido de alcohol, el color del vino aumentará entre 0.1477 y 0.2350 unidades.
Este intervalo de confianza es relativamente estrecho, lo que sugiere que el modelo de regresión es bastante preciso en la estimación del efecto del contenido de alcohol en el color del vino.
Además, el hecho de que el intervalo de confianza no incluya el valor cero sugiere que el efecto del contenido de alcohol en el color del vino es estadísticamente significativo.
En resumen, el intervalo de confianza de 95% sugiere que el contenido de alcohol tiene un efecto estadísticamente significativo en el color del vino, y que el modelo de regresión es bastante preciso en la estimación de este efecto.


```python
# 7. Calcule los residuales y trace un nuevo gráfico de dispersión. Comente, ¿Parece que se verifican los supuestos?
plt.figure()
residuales = y - Yc
plt.scatter(x, residuales, color = 'blue')
plt.xlabel('Alcohol')
plt.ylabel('Color intensity')
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.axhline(y = 0, color = 'gray', linestyle = '--')
#teoria de breush
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
x=sm.add_constant(x)
bp_stat, bp_p_value, _, _ = het_breuschpagan(residuales, x)
print(f'valor-p para breuschpagan: {bp_p_value: 0.4f}\n')
```

Los residuales en el contexto de la regresión lineal se refieren a las diferencias entre los valores observados y los valores predichos por el modelo. En otras palabras, los residuales representan la variabilidad en los datos que no se explica por el modelo.

En el contexto del impacto del alcohol en el color del vino, los residuales podrían ser útiles para:

1. Identificar patrones no lineales: Si los residuales no son aleatorios y muestran un patrón no lineal, podría indicar que la relación entre el alcohol y el color del vino no es lineal.

```python
# 8. Realice la prueba de Shapiro para los residuales y comente el resultado.
from scipy.stats import shapiro
_, valor_p_sh = shapiro(residuales)
print(f'valor-p para shapiro: {valor_p_sh: 0.4f}\n')
```

El valor-p para la prueba de Shapiro es 0.6095.
La prueba de Shapiro es una prueba estadística que se utiliza para evaluar la normalidad de los residuos en un modelo de regresión lineal. La normalidad es una de las suposiciones clave del modelo de regresión lineal.
Un valor-p de 0.6095 sugiere que no hay evidencia estadística significativa para rechazar la hipótesis nula de normalidad. En otras palabras, no hay evidencia de que los residuos no sean normales.
Esto es bueno, ya que la normalidad es una de las suposiciones clave del modelo de regresión lineal. Si los residuos no son normales, puede afectar la precisión de los coeficientes estimados y la validez de las inferencias estadísticas.
En resumen, el valor-p de 0.6095 sugiere que la suposición de normalidad es razonable para este modelo de regresión lineal, lo que nos da más confianza en la precisión de los coeficientes estimados y las inferencias estadísticas.
Ahora que hemos evaluado la homocedasticidad y la normalidad de los residuos, podemos concluir que el modelo de regresión lineal es adecuado para analizar la relación entre el contenido de alcohol y el color del vino.



```python
# 8. teoria de breush
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
x=sm.add_constant(x)
bp_stat, bp_p_value, _, _ = het_breuschpagan(residuales, x)
print(f'valor-p para breuschpagan: {bp_p_value: 0.4f}\n')
```

El valor-p para la prueba de Breusch-Pagan es 0.1467.
La prueba de Breusch-Pagan es una prueba estadística que se utiliza para evaluar la homocedasticidad de los residuos en un modelo de regresión lineal. La homocedasticidad se refiere a la idea de que la varianza de los residuos es constante en todo el rango de valores de la variable independiente.
Un valor-p de 0.1467 sugiere que no hay evidencia estadística significativa para rechazar la hipótesis nula de homocedasticidad. En otras palabras, no hay evidencia de que la varianza de los residuos cambie significativamente en función del contenido de alcohol.

Esto es bueno, ya que la homocedasticidad es una de las suposiciones clave del modelo de regresión lineal. Si la varianza de los residuos no es constante, puede afectar la precisión de los coeficientes estimados y la validez de las inferencias estadísticas.
En resumen, el valor-p de 0.1467 sugiere que la suposición de homocedasticidad es razonable para este modelo de regresión lineal, lo que nos da más confianza en la precisión de los coeficientes estimados y las inferencias estadísticas.
